## 연산자

### 산술 연산자

```python
a = 9
b = 4
print(a, '+', b, '=', a + b)
print(a, '-', b, '=', a - b)
print(a, '*', b, '=', a * b)
print(a, '/', b, '=', a / b)
print(a, '//', b, '=', a // b)
print(a, '%', b, '=', a % b)
print(a, '**', b, '=', a ** b)
```

### 비교(관계) 연산자

```python
a = 3
b = 1
print(a, '==', b, a == b)
print(a, '!=', b, a != b)
print(a, '>', b, a > b)
print(a, '>=', b, a >= b)
print(a, '<', b, a < b)
print(a, '<=', b, a <= b)
```

### 논리 연산자

```python
print(9 % 2 == 1) # T
print(3 >= 1) # T
print((9 % 2 == 1) and (3 >= 1))
```

```python
print(9 % 2 == 0) # F
print(3 >= 1) # T
print((9 % 2 == 0) or (3 >= 1))
```

### 비트 연산자

```python
# Bitwise AND
a = 0b1010
b = 0b1100
print(a, '&', b, a & b)
print(bin(a))
print(bin(b))
print('------')
print(bin(a & b))
```

```python
# Bitwise OR
a = 0b1010
b = 0b1100
print(a, '|', b, a | b)
print(bin(a))
print(bin(b))
print('------')
print(bin(a | b))
```

```python
# Bitwise XOR
a = 0b1010
b = 0b1100
print(a, '^', b, a ^ b)
print(bin(a))
print(bin(b))
print('------')
print(bin(a ^ b))
```

```python
# Bitwise NOT (1의 보수)
a = 0b1010
print(bin(a), bin(~a))
print(a, ~a)
```

```python
# 왼쪽 시프트 (Left Shift)
a = 0b1101001
print(bin(a), bin(a<<2), bin(a<<3))
print(a, a<<2, a<<3)
```

```python
# 오른쪽 시프트 (Right Shift)
a = 0b1101001
print(bin(a), bin(a>>2), bin(a>>3))
print(a, a>>2, a>>3)
```

```python
# 우선 순위
print(False and False or True) # T
print(False and (False or True)) # F
```

## 표준 입출력 함수

### 표준 출력

```python
a = 10
b = 'Peter'
print(b, 'is', a, 'years old.')
print(b+' is '+str(a)+' years old.') # 문자열을 연결하기 위해서 정수를 str 함수를 이용하여 문자열로 바꾼다.
print('%s is %d years old.' % (b, a))
print('{} is {} years old.'.format(b, a))
print('{name} is {age} years old.'.format(name = b, age = a))
```

## 조건문 (if)

```python
# 합격/불합격 판별하기
score = 67
if score >= 50:
  print('합격입니다.')
else:
  print('불합격입니다.')
```

```python
# 짝수/홀수 판별하기
number = int(input("Enter an integer: "))
if number == 0:
  print(number, 'is zero')
elif number % 2 == 0:
  print(number, 'is an even number')
else:
  print(number, 'is an odd number')
```

```python
# 학급 학생인지 판별하기
my_class = ['Rachel', 'Tom', 'James', 'Sarah', 'Julia']
name = input("What is your name? ")
if name in my_class:
  print(name, 'is in my class')
else:
  print(name, 'is not in my class')
```

```python
# pH 농도 값으로 산성 판별하기
# pH = float(input('Enter a pH: '))
pH = 2.1
if pH < 7:
  print('산성 용액입니다.')
  if pH < 3:
    print('강산입니다. 조심하세요.')
elif pH == 7:
  print('중성 용액입니다.')
else:
  print('염기성 용액입니다.')
```

```python
# 성적(A, B, C, D, F) 판별하기
name = input("이름을 입력하세요: ")
score = float(input("점수를 입력하세요: "))
grade = ''
if score < 0:
  print('잘못 입력하셨습니다!')
elif score >= 90:
  if score >= 95:
    grade = 'A+'
  else:
    grade = 'A0'
elif score >= 80:
  grade = 'B+'
elif score >= 70:
  grade = 'C+'
elif score >= 60:
  grade = 'D+'
else:
  grade = 'F'
if grade != '':
  print(name, '학생의 성적은', grade, '입니다.')
```
