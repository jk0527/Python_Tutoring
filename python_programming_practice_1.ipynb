### 딕셔너리 (Dictionary)

```python
# 딕셔너리 함수
print(phone.keys())
print(type(phone.keys()))
print(phone.values())
print(type(phone.values()))
print(phone.items())
print(type(phone.items()))
```

### 집합

```python
a = set([1, 2, 3, 4, 5, 3])
print(a)
print(type(a))
```

```python
# 교집합
a = set([0, 1, 2, 3, 4])
b = set([2, 3, 4, 5, 6])
print(a & b)
print(a.intersection(b))
print(b.intersection(a))
```

```python
# 합집합
a = set([0, 1, 2, 3, 4])
b = set([2, 3, 4, 5, 6])
print(a | b)
print(a.union(b))
print(b.union(a))
```

```python
# 차집합
a = set([0, 1, 2, 3, 4])
b = set([2, 3, 4, 5, 6])
print(a - b) # A - B
print(a.difference(b))
print(b - a) # B - A
print(b.difference(a))
```

```python
# 값 추가하기
a = set([0, 1, 2, 3, 4])
a.add(5)
print(a)
a.update([6, 7])
print(a)
```

```python
# 값 제거하기
a = set([1, 2, 3, 4, 5])
a.remove(4)
print(a)
```
